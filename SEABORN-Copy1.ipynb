```python
import seaborn as sns
sns.set(style="darkgrid")
tendi=sns.load_dataset("tips")
tendi
```

```python
ax=sns.violinplot(x=tendi['total_bill'])
print(ax)
ax=sns.pointplot(x=tendi["total_bill"],y=tendi["smoker"],z=tendi['sex'])
```

```python
ax=sns.pointplot(x=tendi["smoker"],y=tendi["size"])
```

```python
ax=sns.violinplot(x=tendi['smoker'],y=tendi['size'])
```

```python
ax=sns.lineplot(x=tendi['smoker'],y=tendi['size'])
```

```python
ax=sns.boxplot(x=tendi['smoker'],y=tendi['size'])
```

```python
g=sns.PairGrid(tendi)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
ax=g.add_legend()
```
